$$
p^{y_i}(1-p)^{1-y_i}

- Write out the likelihood.

- Bootstrap the sampling density/distribution of $\hat{p}$.

```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import scipy as sp 
```

```python
df = pd.read_csv('metabric.csv')

y = df['Radio Therapy']
sns.histplot(y,stat='proportion')
```

The likelihood is:
$$
L(p) = \prod_{i=1}^n p^{y_i}(1-p)^{1-y_i}
$$

To determine the maximum likelihood with respect to $p$:

1. First, we can find the log-likelihood to make differentiating easier:

$$
\ell(p) = \sum_{i=1}^n y_i \log(p) + (1-y_i) \log(1-p)
$$

2. Then we can take the derivative with respect to $p$

$$
\ell'(p) = \sum_{i=1}^n y_i \frac{1}{p} - (1-y_i) \frac{1}{1-p}
$$

3. Set to zero and solve for $p$.

$$
\sum_{i=1}^n y_i \frac{1}{\hat{p}} - (1-y_i) \frac{1}{1-\hat{p}} = 0
$$

$$
\hat{p} = \dfrac{1}{n} \sum_{i=1}^n y_i.
$$

The MLE, $\hat{p}$ is the sample mean of the $y_i$

```python
y = pd.to_numeric( y.replace({'NO':'0','YES':'1'}) , errors='coerce' )

S = 5000
sample = []

for s in range(S):
    cur = np.mean(y.sample(frac=1,replace=True))
    sample.append(cur)

fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.kdeplot(x=sample,ax=axes[0],label='kde')
axes[0].set_title('Sample KDE')
sns.ecdfplot(x=sample,ax=axes[1],label='ecdf')
axes[1].set_title('Sample ECDF')
plt.show()
```

**Derive the density for this distribution.**
Below is the DESNSITY of the normal distribution:
$$
f(y) = \dfrac{1}{\sqrt{2\pi} \sigma}e^{-\frac{1}{2}\left( \frac{y-\mu}{\sigma} \right)^2}
$$

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

log_price = np.log(ames['price'])

mean = 12
std = 0.4

x = np.linspace(log_price.min(), log_price.max(), 500)
normal_pdf = norm.pdf(x, loc=mean, scale=std)
sns.histplot(log_price, stat='density', kde=True, alpha=0.6)
plt.plot(x, normal_pdf, color='red', linewidth=2)
plt.xlabel("Log Price")
plt.ylabel("Density")
plt.title("Histogram of Log Prices with Normal PDF Overlay")
plt.show()
```

**WRITE OUT THE LIKLIHOOD and SOLVE FOR IT**
The density of the normal distribution, as given above is:
$$
f(y) = \dfrac{1}{\sqrt{2\pi} \sigma}e^{-\frac{1}{2}\left( \frac{y-\mu}{\sigma} \right)^2}
$$
The liklihood of the normal distribution is given by...
$$
L(\mu, \sigma) = \prod_{i=1}^n \dfrac{1}{\sqrt{2\pi} \sigma}e^{-\frac{1}{2}\left( \frac{y_i-\mu}{\sigma} \right)^2}.
$$
We must differentiate by both $\mu$ and $\sigma$ to find the MLEs, but first, to ease our life, we can take the log of the liklihood...
$$\ell(\mu, \sigma)
= \sum_{i=1}^{n} \log\!\left( \frac{1}{\sqrt{2\pi}\,\sigma}
\exp\!\left( -\frac{1}{2} \left( \frac{y_i - \mu}{\sigma} \right)^2 \right)
\right)$$
Final log form...
$$
\ell(\mu, \sigma) = \sum_{i=1}^n \log \left( \dfrac{1}{\sqrt{2\pi}}\right) - \log(\sigma) -\frac{1}{2}\left( \frac{y_i-\mu}{\sigma} \right)^2
$$
Now we can take the derivative with respect to each parameter...
$$
\dfrac{\partial \ell(\mu,\sigma)}{\partial \mu} = \sum_{i=1}^n  \frac{y_i-\mu}{\sigma^2}
$$
$$
\dfrac{\partial \ell(\mu,\sigma)}{\partial \sigma} = \sum_{i=1}^n  \frac{-1}{\sigma} +  \frac{(y_i-\mu)^2}{\sigma^3} 
$$
Now we set each equal to zero and solve for the parameters. Setting the first one equal to zero and solving for mu yields:
$$
\hat{\mu} = \frac{1}{n}\sum_{i=1}^n y_i
$$
setting the second equal to zero and solving for sigma yields:
$$
\hat{\sigma} = \sqrt{ \frac{1}{n} \sum_{i=1}^n (y_i - \hat{\mu})^2 }
$$
This is the maximum likelihood estimate for the parameters of the normal distribution!! 

```python
# Plot the density/distribution for your fitted model and compare it to the KDE/ECDF. Criticize the fit.
    # We basically did this above when we mapped the normal distribution on top of the histogram/PMF but we can do it again below with the CDF and PDF. 
mu_hat = np.mean(log_price)
sigma_hat = np.std(log_price)

grid = np.linspace(log_price.min(), log_price.max(), 500)

cdf = norm.cdf(grid, loc=mu_hat, scale=sigma_hat)
pdf = norm.pdf(grid, loc=mu_hat, scale=sigma_hat)

fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.ecdfplot(x=log_price,ax=axes[0],label='ecdf')
sns.lineplot(x=grid,y=cdf,ax=axes[0],label='cdf',linestyle='--')
axes[0].set_title('ECDF')
sns.kdeplot(x=log_price,ax=axes[1],label='kde')
sns.lineplot(x=grid,y=pdf,ax=axes[1],label='pdf',linestyle='--')
axes[1].set_title('KDE')
plt.show()
```

**FIT**: Generally, our fit looks pretty good. However, our KDE is slightly right skewed compared to the normal distribution. It also has a slightly taller peak than the normal distribution. It is not perfect. This is likely due to outliers (very expensive houses) that are pulling the distribution to the right. The ECDF looks nearly perfect. 

```python
# Bootstrap the sampling density/distribution of your parameters.

# Define reusable function to get MLE estimates
def max_liklihood(data):
    mu = np.mean(data)
    sigma = np.std(data)
    return mu, sigma

# Bootstrap
straps = 1000
sample = [max_liklihood(log_price.sample(frac=1, replace=True)) for i in range(straps)]
mu_samples = [s[0] for s in sample]
sigma_samples = [s[1] for s in sample]

# Plot sampling distributions

# First plot the sampling distribution of mu (ECDF then KDE)
fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.ecdfplot(x=mu_samples,ax=axes[0])
axes[0].set_title('Sampling Distribution of Mu - ECDF')
sns.kdeplot(x=mu_samples,ax=axes[1])
axes[1].set_title('Sampling Distribution of Mu - KDE')
plt.show()

# Now plot the sampling distribution of sigma (ECDF then KDE)
fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.ecdfplot(x=sigma_samples,ax=axes[0])
axes[0].set_title('Sampling Distribution of Sigma - ECDF')
sns.kdeplot(x=sigma_samples,ax=axes[1])
axes[1].set_title('Sampling Distribution of Sigma - KDE')
plt.show()
```

This looks great! This helps us quantity the uncertainty in our parameter estimates.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import poisson
```

```python
sns.histplot(df['Mutation Count'], bins=50)
```

The likelihood is:
$$
L(\lambda) = \prod_{i=1}^n \frac{ \lambda^{y_i} e^{-\lambda} }{y_i!}
$$

To determine the maximum likelihood with respect to $\lambda$:

1. First, we can find the log-likelihood to make differentiating easier:

$$
\ell(\lambda) = \sum_{i=1}^n \log(\lambda^{y_i}) + \log(e^{-\lambda}) - \log(y_i!)
$$

$$
\ell(\lambda) = \sum_{i=1}^n y_i\log(\lambda) - \lambda - \log(y_i!)
$$


2. Then we can take the derivative with respect to $\lambda$

$$
\ell'(\lambda) = \sum_{i=1}^n (\frac{y_i}{\lambda}-1)
$$

3. Set to zero and solve for $p$.

$$
\sum_{i=1}^n (\frac{y_i}{\lambda}-1) = 0
$$

$$
\hat{\lambda} = \dfrac{1}{n} \sum_{i=1}^n y_i.
$$

```python
y = df['Mutation Count'] 
lambda_hat = np.mean(y)
xs = np.arange(0, y.max()+1)
```

```python
poisson_pmf = poisson.pmf(xs, lambda_hat)
poisson_cdf = poisson.cdf(xs, lambda_hat)
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(y, fill=True, label='KDE (empirical)', bw_adjust=0.6)
plt.plot(xs, poisson_pmf, 'o-', label=f'Poisson PMF (={lambda_hat:.2f})')
plt.title("Poisson PMF vs KDE")
plt.xlabel("Mutation Count")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.ecdfplot(y, label='ECDF (empirical)', color='blue')
plt.plot(xs, poisson_cdf, label='Poisson CDF', linewidth=3)
plt.xlabel("Mutation Count")
plt.ylabel("Cumulative Probability")
plt.title("ECDF vs Poisson CDF")
plt.legend()
plt.show()
```

The Poisson PMF is more narrow than the KDE. The Poisson drops off more sharply after the peak. Also, the KDE peak is not as tall as the Poisson PMF.
