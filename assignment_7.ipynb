*What is the expected value of a single die roll?*
- The expected value of a single die roll is the sum of all possible outcomes and their probabilities. A fair die has 6 outcomes {1,2,3,4,5,6}
- $E[X] = \sum_{i=1}^{6} i \times P(X=i)$

*What is the expected winnings of any gamble in European roulette?*:
- The expected winnings of any gamble in European Routlette is based on the number of slots that you bet on. There are 37 slots (0-36), and the payout, if you win, is 36/k-1 where K is the number of slots that you bet on. If you lose, the payout is -1. In both cases, -1 is the dollar you spend to play roulette.
- $P_w = \frac{K}{37}$ and $P_L = 1 - P_w = \frac{37-K}{37}$
- Here is the formula for our expected winnings... 
$$ E = P_w \times \frac{36}{K}-1 + P_L \times -1$$
$$ E = \frac{K}{37} \times (\frac{36}{K}-1) + \frac{37-K}{37} \times -1 $$
$$ E = \frac{K}{37} \times (\frac{36}{K}-1) + \frac{37-K}{37} \times -1 $$
$$ E = \frac{K}{37} \times (\frac{36-K}{K}) - \frac{37-K}{37} $$
$$ E = \frac{1}{37} \times (\frac{36-K}{1}) - \frac{37-K}{37} $$
$$ E = \frac{36-K}{37} - \frac{37-K}{37} $$
$$ E = \frac{-1}{37}$$
$$ E = -.027\text{ CENTS} $$

- *Show that* $\mathbb{E}[a+bX] = a + b\mathbb{E}[X]$
    - We can show this with the properties of integration:
    $$ \mathbb{E}[a + bX] = \int_{-\infty}^{\infty} (a + bX) f_X(x) \, dx $$
    $$ = \int_{-\infty}^{\infty} a f_X(x) \, dx + \int_{-\infty}^{\infty} bX f_X(x) \, dx $$
    $$ = a \int_{-\infty}^{\infty} f_X(x) \, dx + b \int_{-\infty}^{\infty} X f_X(x) \, dx $$
    $$ = a \cdot 1 + b \mathbb{E}[X] $$
    $$ = a + b \mathbb{E}[X] $$


- Show, by example, that $v(\mathbb{E}[X]) \neq \mathbb{E}[v(X)]$, if $v(x) \neq a+bx$. For example, try $v(y) = y^2$ or $v(y)=\sqrt{y}$ with a Bernoulli or uniform or normally distributed random variable. This can be an important thing to remember: The expectation of a transformed random variable is not the transformation of the expected value.
    - See code below...

```python
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sample = np.random.uniform(0, 1, 1000)
```

```python
ev_sample = np.mean(np.sqrt(sample))
print(f"Expected value of v(X): {ev_sample}")
```

```python
sample_ev = np.sqrt(np.mean(sample))
print(f"v(E[X]): {sample_ev}")
```

As we can see above, $E[v(X)]$ is not equal to $v(E[X])$. This demonstrates that the expected value of a nonlinear transformation is not equal to the transformation of the expected value.

```python
sns.lineplot(x=sample, y=np.sqrt(sample))
plt.plot(np.mean(sample), ev_sample, 'o', color='red')
```

- Open the METABRIC data. Make a histogram of 'Ratio Therapy'.
- Let treatment, $T$ be distributed binomial with parameter $p$. Then the contribution to the likelihood for each patient $i$, with $y_i = 0$ for no radiation therapy and $y_i=1$ for radiation therapy, is 
$$
p^{y_i}(1-p)^{1-y_i}

- Write out the likelihood.
- Maximize the likelihood with respect to $p$. What is the MLE, $\hat{p}$?
- Bootstrap the sampling density/distribution of $\hat{p}$.

```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import scipy as sp 
```

```python
df = pd.read_csv('metabric.csv')

y = df['Radio Therapy']
sns.histplot(y,stat='proportion')
```

The likelihood is:
$$
L(p) = \prod_{i=1}^n p^{y_i}(1-p)^{1-y_i}
$$

To determine the maximum likelihood with respect to $p$:

1. First, we can find the log-likelihood to make differentiating easier:

$$
\ell(p) = \sum_{i=1}^n y_i \log(p) + (1-y_i) \log(1-p)
$$

2. Then we can take the derivative with respect to $p$

$$
\ell'(p) = \sum_{i=1}^n y_i \frac{1}{p} - (1-y_i) \frac{1}{1-p}
$$

3. Set to zero and solve for $p$.

$$
\sum_{i=1}^n y_i \frac{1}{\hat{p}} - (1-y_i) \frac{1}{1-\hat{p}} = 0
$$

$$
\hat{p} = \dfrac{1}{n} \sum_{i=1}^n y_i.
$$

The MLE, $\hat{p}$ is the sample mean of the $y_i$

```python
y = pd.to_numeric( y.replace({'NO':'0','YES':'1'}) , errors='coerce' )

S = 5000
sample = []

for s in range(S):
    cur = np.mean(y.sample(frac=1,replace=True))
    sample.append(cur)

fig, axes = plt.subplots(1,2,figsize=(14,5))
sns.kdeplot(x=sample,ax=axes[0],label='kde')
axes[0].set_title('Sample KDE')
sns.ecdfplot(x=sample,ax=axes[1],label='ecdf')
axes[1].set_title('Sample ECDF')
plt.show()
```
